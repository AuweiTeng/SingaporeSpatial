```python
import PyPDF2
import csv
import requests
import pandas as pd
```

```python
reader = PyPDF2.PdfReader('[1]Kimly Group List of Applicable Outlets for Dining Voucher Suppliers _ Kimly Group _ The flavours of Life.pdf')
```

```python
def postaldetect( text_chunk ):
    for i in range(0,len(text_chunk) ):
        if text_chunk[i] == 'S' and text_chunk[i+1:i+1+6].isnumeric():
            return (text_chunk[i:i+1+6])
        elif len(text_chunk) > 1:
            return postaldetect( text_chunk [i+1:] )
        else:
            return 0

# RECURSIVE CODE TO DERIVE POSTAL CODES FROM TEXTS
        
def postal_tabulator ( text_chunk , dict1 ): # PLACES POSTAL CODE INTO A DICTIONARY
    postal = postaldetect ( text_chunk )
    print(postal)
    if postal != 0:
        text_chunk_split = text_chunk.split(postal)
        dict1 [ postal ] = text_chunk_split [0]
        
        #recursively feeds in the remainder text to retrieve the postal codes
        postal_tabulator ( text_chunk_split [1] , dict1 ) 
    else: #the recursive 
        return dict1

def postal_tabulator1 ( text_chunk , df ): # PLACES POSTAL CODE INTO A DATAFRAME
    length = len(df)
    postal = postaldetect ( text_chunk )
    print(postal)
    if postal != 0:
        text_chunk_split = text_chunk.split(postal)
        if len(text_chunk_split) >2:
            print('DoublePostal')
            postal_tabulator1 ( text_chunk_split [2] , df ) 
        else:
            df.at[length,'postal'] = postal
            df.at[length,'address'] = text_chunk_split [0]
            postal_tabulator1 ( text_chunk_split [1] , df ) 
    else:
        return df
```

```python
df1 = pd.DataFrame()
df1['postal'] = ''
df1['address'] = ''

postal_tabulator1(reader.pages[12].extract_text().replace('\n',' ') [92:] , df1 )
```
